# 2. 1次元のHubbard模型を解いてみる

```julia
struct Parm
    t::Float64
    μ::Float64
    U::Float64
    N::Int #SiteNum
    fill::Int #Filling
end
```

```julia
p0 = Parm(0.4, -2.0, 4.0, 100, 100)
```

```julia
function cnstr_state(p::Parm)
    sites = siteinds("Electron",p.N; conserve_qns=true)
    st = [isodd(n) ? "Up" : "Dn" for n in 1:p.fill]
    state = randomMPS(sites, st, 2)

    return sites, state
end
```

```julia
function cnstr_ham(p::Parm, site_ind)
    #空の配列を用意
    ampo = OpSum()

    #hopping項を追加する
    for j=1:p.N-1
        ampo += p.t,"Cup",j,"Cdagup",j+1
        ampo += p.t,"Cdagup",j,"Cup",j+1
        ampo += p.t,"Cdn",j,"Cdagdn",j+1
        ampo += p.t,"Cdagdn",j,"Cdn",j+1
    end

    # 周期境界を付けるなら
    #=
    ampo += p.t,"Cup",1,"Cdagup",p.N
    ampo += p.t,"Cdagup",1,"Cup",p.N
    ampo += p.t,"Cdn",1,"Cdagdn",p.N
    ampo += p.t,"Cdagdn",1,"Cdn",p.N=#

    #onsite項(化学ポテンシャル、Hubbard相互作用項を追加)
    for j in 1:p.N
        ampo += p.μ,"Ntot",j
        ampo += p.U,"Nup",j,"Ndn",j
    end
    #定義された空間上でMPOにする
    H = MPO(ampo,site_ind)
    
    return H
end
```

```julia
function calc_dmrg(p::Parm ,num_sw::Int, mdim::Tuple, cutoff::Float64)
    sites, state = cnstr_state(p)
    H = cnstr_ham(p, sites)
    sweeps = Sweeps(num_sw)
    maxdim!(sweeps, mdim...)
    cutoff!(sweeps, cutoff)

    e, psi = dmrg(H, state, sweeps)

    return e, psi
end
```

```julia
e0, psi0 = calc_dmrg(p0, 30, (10,20,100,100,200), 1E-10)
```
