```julia
using LinearAlgebra
using Plots
```

```julia
w = range(-0.2,0.2,length=201)
w2 = range(-0.4,0.4,length=201)
T = 0.02
beta = 50.0
```

```julia
f0(w0) = 1.0/(1.0+exp(beta*w0))
Ref(w0, eta) = real(1.0/(1.0+exp(beta*(w0+1.0im*eta))))
Imf(w0, eta) = imag(1.0/(1.0+exp(beta*(w0+1.0im*eta))))
edf(w0, eta) = -eta*beta/(1.0+exp(beta*w0))/(1.0+exp(-beta*w0))
df(w0, eta) = beta/(1.0+exp(beta*w0))/(1.0+exp(-beta*w0))
```

```julia
w_l = length(w)
f0_1 =zeros(Float64, w_l)
Ref_1 =zeros(Float64, w_l)
Imf_1 =zeros(Float64, w_l)
edf_1 =zeros(Float64, w_l)
f0_2 =zeros(Float64, w_l)
Ref_2 =zeros(Float64, w_l)
Imf_2 =zeros(Float64, w_l)
edf_2 =zeros(Float64, w_l)
println("I'm ready!")
```

```julia
for i in 1:w_l
    f0_1[i] = f0(w[i])
    Ref_1[i] = Ref(w[i], 0.01)
    Imf_1[i] = Imf(w[i], 0.01)
    edf_1[i] = edf(w[i], 0.01)
    Ref_2[i] = Ref(w[i], 0.02)
    Imf_2[i] = Imf(w[i], 0.02)
    edf_2[i] = edf(w[i], 0.02)
end
```

```julia

Plots.scalefontsizes(1.4)
```

```julia
p1=plot(w, Ref_1, label="Re[f]", xlabel="ω", title="T=0.02, η=0.01", linewidth=3.0)
p1=plot!(w, f0_1, label="f0", linewidth=3.0)
p1=plot!(w, Imf_1, label="Im[f]", linewidth=4.0)
p1=plot!(w, edf_1, label="η*df0", linewidth=2.0, color=:red)
```

```julia
p2=plot(w, Ref_2, label="Re[f]", xlabel="ω", title="T=0.02, η=0.02", linewidth=3.0)
p2=plot!(w, f0_1, label="f0", linewidth=3.0)
p2=plot!(w, Imf_2, label="Im[f]", linewidth=4.0)
p2=plot!(w, edf_2, label="η*df0", linewidth=3.0, color=:red)
```

```julia
A(w0, e, eta) = eta/((w0 - e)^2 + eta^2)/pi
```

```julia
w2_l = length(w2)
df1 = zeros(Float64, w2_l)
df2 = zeros(Float64, w2_l)
A1 = zeros(Float64, w2_l)
A2 = zeros(Float64, w2_l)
println("I'm ready!")
```

```julia
for i in 1:w2_l
    df1[i] = df(w2[i],0.01)
    A1[i] = A(w2[i], 0.1, 0.01)
    A2[i] = A(w2[i], 0.1, 0.1)
end
```

```julia
p3=plot(w2, df1, label="-df/dω", xlabel="ω", title="T=0.02, η=0.01", linewidth=3.0)
p3=plot!(w2, A1, label="A(ω)", linewidth=3.0)
```

```julia
p4=plot(w2, df1, label="-df/dω", xlabel="ω", title="T=0.02, η=0.1", linewidth=3.0)
p4=plot!(w2, A2, label="A(ω)", linewidth=3.0)
```
