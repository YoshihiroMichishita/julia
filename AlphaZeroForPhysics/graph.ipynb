```julia
using LinearAlgebra
using CSV
using DataFrames
using Plots
Plots.scalefontsizes(1.3)
```

```julia
read_data = CSV.read("./PPO_hists_mt14_1124.csv", DataFrame)
```

```julia
p =plot(read_data[:, 1], linewidth=3.0, xaxis=:log, xticks=([1,10,100,1000]), xlabel="iterate", yrange=(0,12), ylabel="max score", gridwidth=2.0,label="trial1", title="AC+PPO", legend=:bottomright)
for it in 2:5
    p=plot!(read_data[:, it], linewidth=3.0,label="trial$(it)")
end

p
savefig(p, "./PPO_hists_mt25_1124.pdf")
```
