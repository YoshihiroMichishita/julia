```julia
df(e::Float64,T::Float64) = 1.0/(1.0+exp(e/T))/(1.0+exp(-e/T))/T
Aw(e::Float64, En::Float64, η::Float64) = η/((e-En)^2 + η^2)/pi
```

```julia
w = collect(-0.5:0.002:0.5)
plot(w, df.(w,0.01),xlabel="ω",ylabel="f(ω), A(ω)",label="-df(ω)", width=4.0,gridlinewidth=2.0,)
plot!(w, Aw.(w,0.1,0.1),label="A(ω)", width=4.0)
```
